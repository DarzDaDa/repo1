```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 生成 N = 100 個來自均值為 0 和標準差為 1 的常態分佈的樣本
N = 100
mu_true = 0
sigma_true = 1
samples = stats.norm.rvs(loc=mu_true, scale=sigma_true, size=N)

# 定義對數聯合概似函數
def log_likelihood(mu, sigma, samples):
    if sigma <= 0:  # 確保標準差為正
        return -np.inf
    N = len(samples)
    return -N * np.log(sigma) - (1 / (2 * sigma**2)) * np.sum((samples - mu)**2)

# 設定 \mu 和 \sigma 的範圍
mu_range = np.linspace(-1, 1, 100)
sigma_range = np.linspace(0.1, 2, 100)

# 計算對數聯合概似函數值
log_likelihood_values = np.zeros((len(sigma_range), len(mu_range)))
for i, sigma in enumerate(sigma_range):
    for j, mu in enumerate(mu_range):
        log_likelihood_values[i, j] = log_likelihood(mu, sigma, samples)

# 繪製等高線圖
X, Y = np.meshgrid(mu_range, sigma_range)
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, log_likelihood_values, levels=1000, cmap='viridis')
plt.colorbar(contour)
plt.title('Contour Plot of Log-Likelihood Function $L(\mu, \sigma|x_1, x_2, \cdots, x_N)$')
plt.xlabel('$\mu$')
plt.ylabel('$\sigma$')
plt.xlim(-1, 1)
plt.ylim(0.1, 2)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# 生成 N = 100 個來自均值為 0 和標準差為 1 的常態分佈的樣本
N = 100
mu_true = 0
sigma_true = 1
samples = stats.norm.rvs(loc=mu_true, scale=sigma_true, size=N)

# 定義聯合對數概似函數
def log_likelihood(mu, sigma, samples):
    if sigma <= 0:  # 確保標準差為正
        return -np.inf
    return np.sum(np.log(stats.norm.pdf(samples, loc=mu, scale=sigma)))

# 定義負對數概似函數
def neg_log_likelihood(params, samples):
    mu, sigma = params
    if sigma <= 0:  # 確保 sigma > 0
        return np.inf
    log_likelihood = np.sum(stats.norm.logpdf(samples, loc=mu, scale=sigma))
    return -log_likelihood  # 這裡放上負號，因為我們需要最小化負的對數概似

# 假設有一些數據
samples = np.random.normal(0, 1, 100)

# 設置初始參數猜測值
initial_guess = [0, 1]

# 使用 minimize 進行優化，最小化負的對數概似函數
result = minimize(neg_log_likelihood, initial_guess, args=(samples,))
mu_est, sigma_est = result.x

# 設定 \mu 和 \sigma 的範圍
mu_range = np.linspace(-1, 1, 100)
sigma_range = np.linspace(0.1, 2, 100)

# 計算聯合對數概似函數值
log_likelihood_values = np.zeros((len(sigma_range), len(mu_range)))
for i, sigma in enumerate(sigma_range):
    for j, mu in enumerate(mu_range):
        log_likelihood_values[i, j] = log_likelihood(mu, sigma, samples)

# 繪製等高線圖
X, Y = np.meshgrid(mu_range, sigma_range)
plt.figure(figsize=(8, 6))
plt.xlim(-1, 1)
plt.ylim(0.1, 2)
contour = plt.contour(X, Y, log_likelihood_values, levels=3000, cmap='viridis', vmin=-1000, vmax=-100)
cbar = plt.colorbar(contour)
cbar.set_label('Log-Likelihood')

# 標註最佳估計的 \mu 和 \sigma
plt.scatter(mu_est, sigma_est, color='red', label='Estimated $(\mu, \sigma)$')




plt.title('Contour Plot of Log-Likelihood Function $L(\mu, \sigma|x_1, x_2, \cdots, x_N)$')
plt.xlabel('$\mu$')
plt.ylabel('$\sigma$')
plt.show()
```
