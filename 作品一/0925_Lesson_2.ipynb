```python
#Generate N samples from chi-square distribution with k degrees of freedom

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


#Parameters
k = 3
N = 1000

#Generate samples
x = np.random.chisquare(k, N)

#Plot histogram
plt.hist(x, bins=50, density=True, alpha=0.6, color='g', edgecolor='black', linewidth=1.2, histtype='bar', label='Histogram', align='mid', rwidth=0.8)

#plot title and labels
plt.title('Chi-square distribution with k = 3')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.legend()
plt.show()

#Plot theoretical PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.chi2.pdf(x, k)
plt.plot(x, p, 'k', linewidth=2)
plt.show()

#Compute mean and variance
mean = np.mean(x)
var = np.var(x)
print('Mean:', mean)
print('Variance:', var)
print('Theoretical mean:', k)
print('Theoretical variance:', 2*k)



```

```python
#generate x-bar N times with n samples from chi-square distribution with k defrees of freedom 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

n = 10
N = 1000
x_bar = np.zeros(N)
for i in range(N):
    x = np.random.chisquare(k, n)
    x_bar[i] = np.mean(x)

#Plot histogram
plt.hist(x_bar, bins=50, density=True, alpha=0.6, color='g', edgecolor='gray', linewidth=1.2, histtype='bar', label='Histogram', align='mid', rwidth=0.8)

#plot title and labels
plt.title('Chi-square distribution with k = 3')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.legend()
plt.show()

```

```python
#Draw a pdf of the t-distribution with 1 to 30 degrees of freedom 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.linspace(-5, 5, 100)
for df in range(1, 31):
    p = stats.t.pdf(x, df)
    plt.plot(x, p, label='df = {}'.format(df), linewidth=2, alpha=0.6, color='g')
plt.legend( loc='upper right', fontsize='small', ncol=3, title='Degrees of freedom')
plt.show()

#plot title and lables
plt.title('t-distribution')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.legend()
plt.show()







```
