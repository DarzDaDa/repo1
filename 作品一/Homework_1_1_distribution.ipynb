<font color=#EF9520> （一）繪製常態分配的 PDF 函數</font>

1. 繪製常態分配不同sigma下可能產生的曲線
2. 繪製常態分配不同mu狀況下可能產生的曲線
<hr>

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))


# First subplot: Normal distribution with different standard deviations
mu = 0
sigma = [0.5, 1, 2, 3, 4, 5]
x = np.linspace(-10, 10, 1000)
for s in sigma:
    y = norm.pdf(x, mu, s)
    ax1.plot(x, y, label=f'$\sigma$ = {s}')
ax1.set_title('Normal Distribution with Different Standard Deviations')
ax1.set_xlabel('x')
ax1.set_ylabel('Probability Density Function')
ax1.grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
ax1.legend()

# Second subplot: Normal distribution with different means
mu = [0, 1, 2, 3, 4, 5]
sigma = 1
x = np.linspace(-5, 10, 1000)
for m in mu:
    y = norm.pdf(x, m, sigma)
    ax2.plot(x, y, label=f'$\mu$ = {m}')
ax2.set_title('Normal Distribution with Different Means')
ax2.set_xlabel('x')
ax2.set_ylabel('Probability Density Function')
ax2.grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
ax2.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

##### 注意事項與討論：

- 以常態中可能改變分配形狀的變數mu和sigma為變數，若一次將變數設置太多，將導致畫出之圖形過於雜亂，因此僅將sigma設於0.5, 1, 2, 3, 4, 5；而mu設於0, 1, 2, 3, 4, 5。
-調整x軸範圍，只集中觀察中間段發生明顯變化之部分
-在固定mu，改變sigma的第一個圖形中，可看到:

>> (1)sigma值越小，在x=0的中心位置，波鋒會越高。

>> (2)sigma值越大，左右尾越厚。

- 在固定sigma，改變mu的第一個圖形中，可看到:

>> (1)常態分配的mu在分配中心，因此mu越大，整個分配會越往右

>> (2)每個常態分配的中心剛好等於其設置之mu，並左右對稱

<font color=#EF9520> （二）繪製t分配的 PDF 函數</font>

1. 繪製不同自由度（從0.1到15，每0.5為一間隔）的t分配pdf函數線圖
2. 將Z-distribution 用紅線描繪出來

<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# 設定不同的自由度
degrees_of_freedom = np.arange(0.1, 15, 0.5)

# 生成 x 值
x = np.linspace(-5, 5, 1000)

# 繪製不同自由度的 t 分佈線圖
plt.figure(figsize=(6, 4))
for df in degrees_of_freedom:
    y = t.pdf(x, df)
    plt.plot(x, y, alpha=0.7, color='#D7A1C5')  # 不添加標籤

# 繪製標準正態分佈 (z 分佈) 的紅線
z = norm.pdf(x)
plt.plot(x, z, label='z distribution', color='red', linewidth=2)

# 添加標題和標籤
plt.title('t-Distribution with Different Degrees of Freedom')
plt.xlabel('Value')
plt.ylabel('Probability Density Function')
plt.grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
plt.legend()  # 只顯示 z 分佈的圖例


# 顯示圖形
plt.show()
```

##### 注意事項與討論：

- t分配的pdf形狀相似於期望值為0變異數為1的常態分布，但較寬且低。
- 圖中t分配粉色部分為不同自由度下的t分配形式，再畫上與其相似的Z分配，並以紅線標記，來對比t分配和Z分配的圖型。
- 自由度越低，中心波鋒越低，且左右尾越厚。
- z分配的形狀大致與t分配相似，但z分配的左右尾較薄。

<font color=#EF9520> （三）繪製卡方分配的 PDF 函數</font>

1. 繪製不同自由度（從3到30，間隔為1）的t分配pdf函數線圖
2. 繪製不同自由度（從30到60，間隔為1）的t分配pdf函數線圖
3. 將第一張圖及第二張圖的線合在一張圖上

<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# 設定不同的自由度範圍
degrees_of_freedom_1 = np.arange(3, 30, 1)
degrees_of_freedom_2 = np.arange(30, 60, 1)

# 生成 x 值
x1 = np.linspace(0, 50, 1000)
x2 = np.linspace(0, 100, 1000)

# 創建 3x1 的子圖
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# 第一張圖: Chi-Square Distribution with Degrees of Freedom 3 to 29
for df in degrees_of_freedom_1:
    y = chi2.pdf(x1, df)
    axs[0].plot(x1, y, label=f'df={df}', alpha=0.7, color='blue')
axs[0].set_title('Chi-Square Distribution with Degrees of Freedom 3 to 29')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Probability Density Function')
axs[0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# 第二張圖: Chi-Square Distribution with Degrees of Freedom 30 to 59
for df in degrees_of_freedom_2:
    y = chi2.pdf(x2, df)
    axs[1].plot(x2, y, label=f'df={df}', alpha=0.7, color='pink')
axs[1].set_title('Chi-Square Distribution with Degrees of Freedom 30 to 59')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Probability Density Function')
axs[1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# 第三張圖: Chi-Square Distribution with Different Degrees of Freedom
for df in degrees_of_freedom_1:
    y = chi2.pdf(x2, df)
    axs[2].plot(x2, y, label=f'df={df}', alpha=0.7, color='blue')
for df in degrees_of_freedom_2:
    y = chi2.pdf(x2, df)
    axs[2].plot(x2, y, label=f'df={df}', alpha=0.7, color='pink')
axs[2].set_title('Chi-Square Distribution with Different Degrees of Freedom')
axs[2].set_xlabel('Value')
axs[2].set_ylabel('Probability Density Function')
axs[2].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# 調整子圖之間的間距並顯示圖形
plt.tight_layout()
plt.show()
```

##### 注意事項與討論：

- 因值域為0到無限大，因此可看到在圖中無論自由度如何，左方皆結束在0之處。
- 在第一張藍色線的圖中，自由度範圍在3到30且間隔為1，為右偏的分配。
- 在第二張粉色線的圖中，自由度範圍在30到60且間隔為1，最後接近常態分配。
- 若只把圖的自由度範圍設定在太小的範圍，就無法觀察到卡方分配在自由度非常大時，將接近常態分配。
- 前兩張圖的y軸範圍不同，但在分開兩張圖中很可能因比例尺而誤判圖型樣貌，因此第三張圖將前兩張線圖放在第三張圖，並以藍色和粉色作為分別。
- 第三張圖中可看出自由度越小，波鋒越低。
- 卡方分配中隨著自由度越大，圖型看起來越扁平。

<font color=#EF9520> （四）繪製F分配的 PDF 函數</font>

1. 繪製不同自由度（df1:3到30，每3為一間隔；df2:從3到30，每3為一間隔）的F分配pdf線圖
2. 繪製不同自由度（df1:30到120，每5為一間隔；df2:從30到120，每5為一間隔）的F分配pdf線圖，並標示df1=df2=100為紅線


<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f
import itertools

# 設定不同的自由度範圍
df1_range_1 = np.arange(3, 30, 3)
df2_range_1 = np.arange(3, 30, 3)
df1_range_2 = np.arange(30, 120, 5)
df2_range_2 = np.arange(30, 120, 5)

# 生成所有可能的自由度組合
degrees_of_freedom_1 = list(itertools.product(df1_range_1, df2_range_1))
degrees_of_freedom_2 = list(itertools.product(df1_range_2, df2_range_2))

# 生成 x 值
x1 = np.linspace(0, 5, 1000)
x2 = np.linspace(0, 5, 1000)

# 創建 2x1 的子圖
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# 第一張圖: F-Distribution with Different Degrees of Freedom (3 to 30)
for df1, df2 in degrees_of_freedom_1:
    y = f.pdf(x1, df1, df2)
    axs[0].plot(x1, y, label=f'df1={df1}, df2={df2}', alpha=0.7, color='#549ED7', linewidth=0.8)
axs[0].set_title('F-Distribution with Different Degrees of Freedom (3 to 30)')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Probability Density Function')
axs[0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
axs[0].set_xlim([0, 5])
axs[0].set_ylim([0, 1.2])

# 第二張圖: F-Distribution with Different Degrees of Freedom (30 to 120)
for df1, df2 in degrees_of_freedom_2:
    if df1 == 100 and df2 == 100:
        axs[1].plot(x2, f.pdf(x2, df1, df2), label=f'df1={df1}, df2={df2}', alpha=0.5, color='red', linewidth=5)
    else:
        axs[1].plot(x2, f.pdf(x2, df1, df2), alpha=0.5, color='#4CC053', linewidth=0.8)
axs[1].set_title('F-Distribution with Different Degrees of Freedom (30 to 120)')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Probability Density Function')
axs[1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
axs[1].set_xlim([0, 5])
axs[1].set_ylim([0, 2.5])

# 只顯示紅色線條的圖例
red_line = plt.Line2D([0], [0], color='red', linewidth=5, label='df1=100, df2=100')
axs[1].legend(handles=[red_line])

# 調整子圖之間的間距並顯示圖形
plt.tight_layout()
plt.show()
```

##### 注意事項與討論：

- 使用for迴圈為df1和df2創造不同組合，以看出更多F分配圖型樣態。
- F分配的範圍在0到無限大之間，與卡方分配相同，因此x軸設置最低為0，而圖型右方將持續延續下去，無特別觀察之必要。
- 第一張圖的自由度範圍將兩者都設於3-30之間，先觀察自由度較小之圖型樣態；第二張圖自由度範圍設於30-120之間，雖然線條密度較密集，但此張圖目的在於觀察F分配是否也會像卡方分配般，自由度越大越接近常態分配。
- 從兩張圖中可歸納上述問題之結論：f分配不如卡方分配般對自由度敏感，即便自由度再大，圖型樣態也僅是像sigma=0, mu=1之常態分配，波鋒中心越來越偏向1。
- F分配為右偏分佈，自由度越大越接近常態。
- df1/df2之數字若越大，圖型越扁平。
- 第二張圖紅線標示之df1=df2=100長得最像常態分配之圖型。

<font color=#EF9520> （五）繪製Beta分配的 PDF 函數</font>

1. 繪製alpha 固定為10，beta範圍變動在1到50且間隔為1的Beta分配線圖
2. 繪製alpha 範圍變動在1到50且間隔為1的Beta分配圖，beta固定為10
3. 繪製alpha = beta（範圍在1到30且間隔為2) 之beta 分配圖
4. 繪製alpha 、beta 皆變動（範圍在1到10且間隔為2)之beta分配圖


<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
import itertools

# 設定 alpha 固定值和 beta 的範圍
alpha_fixed = 10
beta_range = np.arange(1, 50, 1)

# 生成所有可能的參數組合
alpha_beta_pairs = [(alpha_fixed, beta_val) for beta_val in beta_range]

# 生成 x 值
x = np.linspace(0, 1, 1000)

# 創建 2x2 的子圖
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# 第一張圖: Beta Distribution with Different Beta Parameters (Alpha = 10)
for alpha, beta_val in alpha_beta_pairs:
    y = beta.pdf(x, alpha, beta_val)
    axs[0, 0].plot(x, y, alpha=0.7, color='#C5C950')
axs[0, 0].set_title('Beta Distribution with Different Beta Parameters (Alpha = 10)')
axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Probability Density Function')
axs[0, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# 設定 alpha 的範圍和 beta 固定值
alpha_range = np.arange(1, 50, 1)
beta_fixed = 10

# 第二張圖: Beta Distribution with Different Alpha Parameters (Beta = 10)
for alpha in alpha_range:
    y = beta.pdf(x, alpha, beta_fixed)
    axs[0, 1].plot(x, y, alpha=0.7, color='#C5C950')
axs[0, 1].set_title('Beta Distribution with Different Alpha Parameters (Beta = 10)')
axs[0, 1].set_xlabel('Value')
axs[0, 1].set_ylabel('Probability Density Function')
axs[0, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# 設定 alpha 和 beta 的範圍
alpha_beta_range = np.arange(1, 30, 2)

# 第三張圖: Beta Distribution with Alpha = Beta
for alpha_beta in alpha_beta_range:
    y = beta.pdf(x, alpha_beta, alpha_beta)
    axs[1, 0].plot(x, y, alpha=0.8, linewidth=1.3, color='#C5C950')
axs[1, 0].set_title('Beta Distribution with Alpha = Beta')
axs[1, 0].set_xlabel('Value')
axs[1, 0].set_ylabel('Probability Density Function')
axs[1, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# 設定 alpha 和 beta 的範圍
alpha_range = np.arange(1, 10, 2)
beta_range = np.arange(1, 10, 2)

# 生成所有可能的參數組合
alpha_beta_pairs = list(itertools.product(alpha_range, beta_range))

# 第四張圖: Beta Distribution with Different Alpha and Beta Parameters
for alpha, beta_val in alpha_beta_pairs:
    y = beta.pdf(x, alpha, beta_val)
    axs[1, 1].plot(x, y, alpha=0.8, color='#C5C950', linewidth=1.3)
axs[1, 1].set_title('Beta Distribution with Different Alpha and Beta Parameters')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Probability Density Function')
axs[1, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# 調整子圖之間的間距並顯示圖形
plt.tight_layout()
plt.show()
```

##### 注意事項與討論：

- 在Beta分配中有alpha, beta兩個變數，因此做四張圖：固定alpha，改變beta；固定beta，改變alpha；alpha=beta；alpha和beta同時變動，以觀察四種圖形的不同樣貌。
- beta分配的值域在0到1之間，因此x軸也設置在0-1。
- 第一張圖與第二張圖皆固定alpha 或 beta其一為10，並改動另一參數值，可發現兩張圖放在一起時看起來是對稱的，若是固定alpha，改變beta，則分佈右偏；若是固定beta，改變alpha，則分佈左偏
- 從第三張圖及第四張圖中可看出當alpha=beta=1時，beta分配為一橫直線。
- 第三張圖中可看出，當alpha=beta時，分配的形狀似常態分配。
