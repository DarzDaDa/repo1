```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def remove_outliers(data, threshold=1.5):
    """移除超過指定四分位距倍數的離群值"""
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    return data[(data > lower_bound) & (data < upper_bound)]

def plot_distributions(df, sample_size, ax, title_suffix):
    """繪製卡方分配的直方圖、盒鬚圖、常態機率圖和經驗CDF圖"""
    data = np.random.chisquare(df, sample_size)
    data = remove_outliers(data)

    # 直方圖
    ax[0, 0].hist(data, bins=30, color='lightblue', edgecolor='black')
    ax[0, 0].set_title(f'Histogram of Chi-Square Distributed Data (df={df}, N={sample_size})')
    ax[0, 0].set_xlabel('Value')
    ax[0, 0].set_ylabel('Frequency')
    ax[0, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 盒鬚圖
    ax[0, 1].boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    ax[0, 1].set_title(f'Boxplot of Chi-Square Distributed Data (df={df}, N={sample_size})')
    ax[0, 1].set_xlabel('Value')
    ax[0, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 常態機率圖
    stats.probplot(data, dist="norm", plot=ax[1, 0])
    ax[1, 0].get_lines()[0].set_color('blue')
    ax[1, 0].set_title(f'Normal Probability Plot (df={df}, N={sample_size})')
    ax[1, 0].set_xlabel('Theoretical Quantiles')
    ax[1, 0].set_ylabel('Sample Quantiles')
    ax[1, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 經驗CDF圖
    sorted_data = np.sort(data)
    ecdf = np.arange(1, len(data) + 1) / len(data)
    ax[1, 1].step(sorted_data, ecdf, where='post', color='blue', label='Empirical CDF')
    ax[1, 1].set_title(f'Empirical CDF of Chi-Square Distributed Data (df={df}, N={sample_size})')
    ax[1, 1].set_xlabel('Value')
    ax[1, 1].set_ylabel('ECDF')
    ax[1, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
    ax[1, 1].legend()

# 設定卡方分配的參數
df = 10  # 自由度

# 創建 2x2 的子圖
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df, 100, axs, 'N=100')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df, 1000, axs, 'N=1000')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df, 10000, axs, 'N=10000')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df, 100000, axs, 'N=100000')
plt.tight_layout()
plt.show()
```

##### 注意事項與討論：

- 直方圖：N=100時看不出往左或右偏移，而隨著樣本數越大可看出分配右偏之現象。
- 盒狀圖：中位數雖接近自由度10，但即便樣本數再大，也無法完全趨近10，較靠近盒子的左方，且上鬚較下鬚長，右側離群值也較左側多，係因卡方分配是一右偏的分配。
- QQ plot：右側大幅偏離直線，呈一橫直線狀；左側低分位數的點會明顯低於對角線，表明數據的左側尾部比正態分佈短。
- ECDF：在左側會先快速上升，因卡方分配為右偏分配，大部分的數據值較小，又因卡方分配又尾較長，ECDF上升趨勢慢，最右上方平緩，尾部數據稀鬆。

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def remove_outliers(data, threshold=1.5):
    """移除超過指定四分位距倍數的離群值"""
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    return data[(data > lower_bound) & (data < upper_bound)]

def plot_distributions(df1, df2, sample_size, ax, title_suffix):
    """繪製F分配的直方圖、盒鬚圖、常態機率圖和經驗CDF圖"""
    data = np.random.f(df1, df2, sample_size)
    data = remove_outliers(data)

    # 直方圖
    ax[0, 0].hist(data, bins=30, color='lightblue', edgecolor='black')
    ax[0, 0].set_title(f'Histogram of F-Distributed Data (df1={df1}, df2={df2}, N={sample_size})')
    ax[0, 0].set_xlabel('Value')
    ax[0, 0].set_ylabel('Frequency')
    ax[0, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 盒鬚圖
    ax[0, 1].boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    ax[0, 1].set_title(f'Boxplot of F-Distributed Data (df1={df1}, df2={df2}, N={sample_size})')
    ax[0, 1].set_xlabel('Value')
    ax[0, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 常態機率圖
    stats.probplot(data, dist="f", sparams=(df1, df2), plot=ax[1, 0])
    ax[1, 0].get_lines()[0].set_color('blue')
    ax[1, 0].set_title(f'Normal Probability Plot (df1={df1}, df2={df2}, N={sample_size})')
    ax[1, 0].set_xlabel('Theoretical Quantiles')
    ax[1, 0].set_ylabel('Sample Quantiles')
    ax[1, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 經驗CDF圖
    sorted_data = np.sort(data)
    ecdf = np.arange(1, len(data) + 1) / len(data)
    ax[1, 1].step(sorted_data, ecdf, where='post', color='blue', label='Empirical CDF')
    ax[1, 1].set_title(f'Empirical CDF of F-Distributed Data (df1={df1}, df2={df2}, N={sample_size})')
    ax[1, 1].set_xlabel('Value')
    ax[1, 1].set_ylabel('ECDF')
    ax[1, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
    ax[1, 1].legend()

# 設定 F 分配的參數
df1 = 5  # 自由度1
df2 = 10  # 自由度2

# 創建 2x2 的子圖
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df1, df2, 100, axs, 'N=100')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df1, df2, 1000, axs, 'N=1000')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df1, df2, 10000, axs, 'N=10000')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(df1, df2, 100000, axs, 'N=100000')
plt.tight_layout()
plt.show()
```

##### 注意事項與討論：

- 直方圖：在N=100無明顯分配趨勢，但在N=10000中可看出F分配為一右偏分配。
- 盒狀圖：中位數與自由度似乎無太大關係，只可看出中位數偏向盒子左方，且上鬚較下鬚長，且右尾較多異常值，這些都是因為F分配是一右偏的分配。
- QQ plot：右上方偏離常態許多，近乎成一條橫直線，表示比常態分配尾部長，可看出F分配是一右偏分配，左下方數據則比常態分配小。
- ECDF：ECDF在左側一開始便快速累積，係因F分配為右偏分配，大部分數據較低而快速累積，到了右方則開始緩慢累積，右側曲線看起來較為平緩。

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def remove_outliers(data, threshold=1.5):
    """移除超過指定四分位距倍數的離群值"""
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    return data[(data > lower_bound) & (data < upper_bound)]

def plot_distributions(alpha, beta, sample_size, ax, title_suffix):
    """繪製Beta分配的直方圖、盒鬚圖、常態機率圖和經驗CDF圖"""
    data = np.random.beta(alpha, beta, sample_size)
    data = remove_outliers(data)

    # 直方圖
    ax[0, 0].hist(data, bins=30, color='lightblue', edgecolor='black')
    ax[0, 0].set_title(f'Histogram of Beta-Distributed Data (alpha={alpha}, beta={beta}, N={sample_size})')
    ax[0, 0].set_xlabel('Value')
    ax[0, 0].set_ylabel('Frequency')
    ax[0, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 盒鬚圖
    ax[0, 1].boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    ax[0, 1].set_title(f'Boxplot of Beta-Distributed Data (alpha={alpha}, beta={beta}, N={sample_size})')
    ax[0, 1].set_xlabel('Value')
    ax[0, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 常態機率圖
    stats.probplot(data, dist="norm", plot=ax[1, 0])
    ax[1, 0].get_lines()[0].set_color('blue')
    ax[1, 0].set_title(f'Normal Probability Plot (alpha={alpha}, beta={beta}, N={sample_size})')
    ax[1, 0].set_xlabel('Theoretical Quantiles')
    ax[1, 0].set_ylabel('Sample Quantiles')
    ax[1, 0].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

    # 經驗CDF圖
    sorted_data = np.sort(data)
    ecdf = np.arange(1, len(data) + 1) / len(data)
    ax[1, 1].step(sorted_data, ecdf, where='post', color='blue', label='Empirical CDF')
    ax[1, 1].set_title(f'Empirical CDF of Beta-Distributed Data (alpha={alpha}, beta={beta}, N={sample_size})')
    ax[1, 1].set_xlabel('Value')
    ax[1, 1].set_ylabel('ECDF')
    ax[1, 1].grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
    ax[1, 1].legend()

# 設定 Beta 分配的參數
alpha = 2  # alpha 參數
beta = 10  # beta 參數

# 創建 2x2 的子圖
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(alpha, beta, 100, axs, 'N=100')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(alpha, beta, 1000, axs, 'N=1000')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(alpha, beta, 10000, axs, 'N=10000')
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
plot_distributions(alpha, beta, 100000, axs, 'N=100000')
plt.tight_layout()
plt.show()
```

##### 注意事項與討論：

- 直方圖：在N=100中可以些微看出Beta分配是一右偏分配，不過在N=10000時會更加明顯，從前一頁的作業中我們探討不同變數的alpha和beta會生成何種型態，因此例將參數固定alpha＝2, beta=10，因此此處的beta分配將會是右偏型態。
- 盒狀圖：中位數靠近盒子左側，且上鬚大於下鬚，係因此處alpha < beta ，因此右偏，又在右側出現較多離群值。
- QQ plot：右上方偏離常態呈現一橫線狀，左下方也偏離常態，整題呈現右偏分佈。
- ECDF：左側快速累積，因分配右偏，數據多集中在左側，到了右上方累積變緩慢，因Beta 分配的右尾較短，較少的數據點分佈在高值區域。
