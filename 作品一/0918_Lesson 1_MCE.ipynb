```python
# generate N samples from a normal distribution with mean 0 and standard deviation 1 and plot the histogram of the samples and the pdf of the normal distribution
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

N = 1000
samples = np.random.normal(0, 1, N)
plt.hist(samples, bins= 50, density=True)
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, 'r')
plt.show()

# generate N samples from a normal distribution with mean 0 and standard deviation 1 and plot the histogram of the samples and the pdf of the normal distribution using seaborn library 
import seaborn as sns
sns.set()
N = 1000
samples = np.random.normal(0, 1, N)
sns.histplot(samples, bins= 50, kde=True)
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, 'r')
plt.show()








```

```python
# generate N samples from a normal distribution with mean 0 and standard deviation 1 and plot the histogram of the samples and the pdf of the normal distribution Plot the PDF of the normal distribution with mean 0 and standard deviation 1
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

N = 100000
samples = np.random.normal(0, 1, N)
plt.hist(samples, bins= 30, density=True, edgecolor='black', alpha=0.6, color='skyblue', linewidth=1.0, histtype='bar', rwidth=0.8, align='mid')
# plot title and labels 
plt.title('Histogram PDF of normal distribution samples')
plt.xlabel('Value')
plt.ylabel('Density')
x = np.linspace(-5, 5, 300)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, 'r', linewidth=2, label='PDF', alpha=0.8)
plt.legend( loc='upper right', fontsize='large', shadow=True)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.5, color='gray')
plt.show()



```
